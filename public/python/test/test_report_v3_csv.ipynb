```python
import pandas as pd
import matplotlib.pyplot as plt

# Langkah 1: Baca data dari CSV
csv_filename = 'simplified_report_counts_20240715_140235.csv'
report_df = pd.read_csv(csv_filename)

# Filter data untuk testid 24-44 dan hanya status FAILED
filtered_report_df = report_df[(report_df['testid'] >= 24) & (report_df['testid'] <= 44) & (report_df['status'] == 'FAILED')]

# Tampilkan data yang difilter
print("Data filtered for testid 24-44 and status FAILED:")
print(filtered_report_df)

# Langkah 2: Visualisasi menggunakan bar chart
plt.figure(figsize=(12, 6))
plt.bar(filtered_report_df['testid'], filtered_report_df['count'], color='red', label='FAILED')

plt.title('Count of FAILED Status for Test ID 24-44')
plt.xlabel('Test ID')
plt.ylabel('Count')
plt.xticks(filtered_report_df['testid'].unique())
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
